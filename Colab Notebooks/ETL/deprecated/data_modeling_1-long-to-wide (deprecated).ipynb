```python
sidra_df = pd.read_csv('/content/drive/MyDrive/Dados/Dados_SIDRA/table_full.csv')
sidra_df.columns
```

```python
sidra_df = sidra_df.reset_index()
sidra_df.head(10)
```

```python
cols = sidra_df['ano'].unique()
df_wide=pd.pivot_table(sidra_df, columns = 'ano',values = 'valor',aggfunc=np.sum,
                        index = ['cidade', 'cod_cidade', 'nome_rgi', 'cod_rgi', 'cod_rgint',
                                'nome_rgint', 'tipo_producao', 'cod_variavel', 'variavel',
                                'cod_produto', 'produto', 'cod_metrica', 'metrica'], fill_value = 0)
df_wide
```

```python
df_wide = df_wide.reset_index()
```

```python
df_wide = df_wide.reset_index(drop=True)
df_wide = df_wide.replace('...', 0)
df_wide = df_wide.replace('-', 0)
df_wide = df_wide.replace('--', 0)
df_wide = df_wide.replace('......', 0)

df_wide
```

```python
columns = list(range(1974, 2022))
df_wide[columns] = df_wide[columns].replace(to_replace=None, regex={'[^0-9]':0})
df_wide[columns] = df_wide[columns].astype('int64')
```

```python
df_wide.to_csv('/content/drive/MyDrive/Dados/Dados_SIDRA/table_full-formated.csv', index = False)
```
